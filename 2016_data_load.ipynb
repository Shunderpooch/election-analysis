```python
import pandas as pd
import os
```

```python
# Pull data from https://digitalcommons.bucknell.edu/election_data/1/#attach_additional_files and put it in the /data folder of your project
data_path = "data/Finalized_2016_State_Primary_and_Caucus_Results/" 
file_names_2016 = [f for f in os.listdir(data_path) if os.path.isfile(os.path.join(data_path, f))]
```

```python
print(file_names_2016)
```

```python
# Create a dict with all the state names and file paths
states_dict = {}

for state in file_names_2016:
    state_name = ""
    try:
        state_name = state[:state.index(" Primary")]
    except ValueError:
        state_name = state[:state.index(" Caucus")]
    states_dict[state_name] = {"state_file_name" : state}
```

```python
print(states_dict)
```

```python
for state in states_dict:
    print(state)
    states_dict[state]['df'] = pd.read_csv(os.path.join(data_path, states_dict[state]["state_file_name"]), encoding = "ISO-8859-1")
```

```python
states_dict["Alabama"]["df"]["state"] = 'Alabama'
states_dict["Alaska"]["df"]["state"] = "Alaska"
result_df = pd.concat([states_dict["Alabama"]["df"], states_dict["Alaska"]["df"]])
display(result_df)
```

```python
final_df = None
for state in states_dict:
    states_dict[state]["df"]["state"] = state
    if final_df is not None:
        final_df = pd.concat([final_df, states_dict[state]["df"]])
    else:
        final_df = states_dict[state]["df"]

display(final_df)
```
